### Feature Selection and Model building

- Separate the features and target and then split the train data into train and validation set.
- Apply different models of your choice and  then predict on the validation data and find the `accuracy_score` for this prediction.
- Try improving upon the `accuracy_score` using different feature selection techniques like wrapper methods, PCA and try using hyperparameter tuning to find the parameters that give the best accuracy.

```python
# Code Starts here

X = train.drop('Cover_Type',1)
y = train['Cover_Type']

X_train,X_test,y_train,y_test=train_test_split(X,y ,test_size=0.3,random_state=0)
regressor=LogisticRegression()

# fit the model
regressor.fit(X_train,y_train)

# predict the result
y_pred =regressor.predict(X_test)


# Calculate roc_score
roc = accuracy_score(y_test, y_pred)

#print roc
print(roc)

# Code ends here
```

```python
nof_list=[5,10,15,20,25,30,35,40,45,50,55]
high_score=0

#Variable to store the optimum features
nof=0

#Loop to select the optimum features
for n in nof_list:
    rfe = RFE(regressor, n)
    X_train_rfe = rfe.fit_transform(X_train, y_train)
    X_test_rfe = rfe.transform(X_test)
    regressor.fit(X_train_rfe,y_train)
    if regressor.score(X_test_rfe,y_test)>high_score:
        high_score=regressor.score(X_test_rfe,y_test)
        nof=n


#Printing the no. features with the highest score along with the highest score
print("No. of features=",nof, "gives the best score=",high_score)
```

```python
scaler=StandardScaler()

#Scaling the features
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#Initialising PCA
pca = PCA(n_components=32)

#Transforming the features
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca=pca.transform(X_test_scaled)

#Initialising the model
model= LogisticRegression()

#Fitting the model
model.fit(X_train_pca,y_train)

#Scoring the model
pca_score=model.score(X_test_pca,y_test)
print(pca_score)
```

```python
pipe = Pipeline([('classifier' , RandomForestClassifier())])
param_grid = [
    {'classifier' : [LogisticRegression()],
     'classifier__penalty' : ['l1', 'l2'],
    'classifier__C' : np.logspace(-4, 4, 20),
    'classifier__solver' : ['liblinear']},
    {'classifier' : [RandomForestClassifier()],
    'classifier__n_estimators' : list(range(10,101,10)),
    'classifier__max_features' : list(range(6,32,5))}
]

# Create grid search object

grid = GridSearchCV(pipe,param_grid = param_grid, cv = 2, verbose=True)

grid_results = grid.fit(X_train, y_train)
#print(grid_results.best_score_, grid_results.best_params_)
# Summarize the results in a readable format
print("Best: {0}, using {1}".format(grid_results.best_score_, grid_results.best_params_))
```
